```python
#Import needed libraries and get data
%matplotlib inline

import thinkstats2
import thinkplot
import pandas as pd
import numpy as np

titanic_df = pd.read_csv('train.csv')
titanic_df
```

```python
titanic_df.describe()
```

Looking at the above data, a few factors immediately stand out to me to visualize. These include pclass, sex, age, and fare. I also thought it could be potentially useful or illuminating to create a visualization of passenger survival rate by cabin number, but there does not seem to be enough data to be able to gain any information. However, for each of the factors I will be analyzing I would like to keep the data size the same, so I will filter out those passengers who have NaN in any of the columns I chose to analyze. 

```python
titanic_df = titanic_df.dropna(subset=['Survived','Pclass','Sex','Age','Fare'])
titanic_df
```

```python
males = titanic_df[titanic_df.Sex == 'male']
females = titanic_df[titanic_df.Sex == 'female']

male_sur = (males[males.Survived == 1])
female_sur = (females[females.Survived == 1])

males_survived_perc = float(len(male_sur)) / float(len(males)) * 100
females_survived_perc = float(len(female_sur)) / float(len(females)) * 100 

print 'total males: ', len(males), ' num survived: ', len(male_sur), ' survival rate: %', males_survived_perc
print 'total females: ', len(females), ' num survived: ', len(female_sur), ' survival rate: %', females_survived_perc
```

It is clear from these survival rates that a higher percentage of females survived; however, there are nearly twice as many men as women, so the distribution is not proportionally represented by survival rate. Therefore, it will be helpful to create a graphic representation. 

```python
hist = thinkstats2.Hist(males.Survived)
hist1 = thinkstats2.Hist(females.Survived)
thinkplot.PrePlot(2)
thinkplot.Hist(hist, align='left', width=0.5, label='males')
thinkplot.Hist(hist1, align='right', width=0.5, label='females')
thinkplot.Show(xlabel='survived', ylabel='amount')
```

From this, it is easier to see the relationship between survival rate and sex. Overall, females had a higher chance of surviving, which makes sense because when the Titanic was being evacuated there was an emphasis placed on getting the women and children on lifeboats first. 

```python
hist = thinkstats2.Hist(male_sur.Age)
thinkplot.Hist(hist, label='males')
thinkplot.Show(xlabel='age', ylabel='amount')
```

From this visualization of surviving males, it can be seen that the majority of males who survived were between the ages of mid-20's to mid-30's. Interestingly, there are also 2 children under the age of 10, supporting the theory that children were a higher priority. However, this conclusion cannot be made without visualizing the ages of males who unfortunately did not survive. 

```python
male_not = males[males.Survived == 0]
hist = thinkstats2.Hist(male_not.Age)
thinkplot.Hist(hist, label='males')
thinkplot.Show(xlabel='age', ylabel='amount')
```

Here, there is a higher distribution of males who did not survive from the ages of the teens to early 30s, with a spike around the late 30's. 

```python
hist = thinkstats2.Hist(female_sur.Age)
thinkplot.Hist(hist, label='females')
thinkplot.Show(xlabel='age', ylabel='amount')
```

For women, the highest survival is around the ages of 20's to 30's, with a spike around age 5. 

```python
female_not = females[females.Survived == 0]
hist = thinkstats2.Hist(female_not.Age)
thinkplot.Hist(hist, label='females')
thinkplot.Show(xlabel='age', ylabel='amount')
```

The first thing that I notice about this graph is that the y-axis range is noticeably smaller than the survival y-axis range, corroborating the fact that %70 of women in this data set survived. This highest numbers here were very small children, possibly because of weaker immune systems, and between the ages of 15 to 30. 

The other area that potentially holds useful information is visualizing survival rate by passenger class and fare, to see if there was a higher priority for first class or higher paying passengers, or if it didn't make a huge difference overall. 

```python
hist = thinkstats2.Hist(titanic_df[titanic_df.Pclass == 1].Survived)
hist1 = thinkstats2.Hist(titanic_df[titanic_df.Pclass == 2].Survived)
hist2 = thinkstats2.Hist(titanic_df[titanic_df.Pclass == 3].Survived)
thinkplot.PrePlot(3)
thinkplot.Hist(hist, align='left', width=0.3, label='Pclass 1')
thinkplot.Hist(hist1, align='center', width=0.3, label='Pclass 2')
thinkplot.Hist(hist2, align='right', width=0.3, label='Pclass 3')
thinkplot.Show(xlabel='survived', ylabel='amount')
```

From this, two things can be seen. The first, that Pclass 1 is the smallest quantity of passengers, and that the majority were in class 3. However, the second piece of information that is seen is that the majority who survived were from Pclass 1. This indicates that either there was a higher priority placed on these passengers due to the money they had paid, or that there is a higher distribution of females and passengers between 25-35 in Pclass 1. 

```python
bins = np.arange(0, 40, 5)
indices = np.digitize(titanic_df.Age, bins)
groups = titanic_df.groupby(indices)

for i, group in groups:
    print(i, len(group))
    age = [group.Age.mean() for i, group in groups]
    cdfs = [thinkstats2.Cdf(group.Pclass) for i, group in groups]
    
for percent in [75, 50, 25]:
    Pclasses = [cdf.Percentile(percent) for cdf in cdfs]
    label = '%dth' % percent
    thinkplot.Plot(age, Pclasses, label=label)
```

Very surprisingly, it appears that the majority of passengers in class 1 appear to be older. Thus, we will look at distribution of gender in each class. 

```python
bins = np.arange(0, 40, 5)
indices = np.digitize(males.Age, bins)
groups = males.groupby(indices)

for i, group in groups:
    print(i, len(group))
    age = [group.Age.mean() for i, group in groups]
    cdfs = [thinkstats2.Cdf(group.Pclass) for i, group in groups]
    
for percent in [75, 50, 25]:
    Pclasses = [cdf.Percentile(percent) for cdf in cdfs]
    label = '%dth' % percent
    thinkplot.Plot(age, Pclasses, label=label)
```

```python
bins = np.arange(0, 40, 5)
indices = np.digitize(females.Age, bins)
groups = females.groupby(indices)

for i, group in groups:
    print(i, len(group))
    age = [group.Age.mean() for i, group in groups]
    cdfs = [thinkstats2.Cdf(group.Pclass) for i, group in groups]
    
for percent in [75, 50, 25]:
    Pclasses = [cdf.Percentile(percent) for cdf in cdfs]
    label = '%dth' % percent
    thinkplot.Plot(age, Pclasses, label=label)
```

From these plots, it seems that my hypothesis was correct - more females are among classes 1 and 2, particularly among 1. There are very few women among the third class, potentially because of the ages - many of the women could be unmarried, and would probably be less likely to travel alone in class 3.
